```python

import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 3, 400)


y1 = 2 - 1.0001 * x
y2 = 2 - x


plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$1.0001x + y = 2$', color='blue')
plt.plot(x, y2, label=r'$x + y = 2$', color='red')


plt.title('Grfico del sistema de ecuaciones')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.xlim(0, 3)
plt.ylim(-1, 3)
plt.show()
```
