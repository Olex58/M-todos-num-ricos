```python
import numpy as np

A = np.array([
    [52/100, 20/100, 25/100],
    [30/100, 50/100, 20/100],
    [18/100, 30/100, 55/100]
])

B = np.array([4800, 5810, 5690])

try:
    Ainv = np.linalg.inv(A)
    C=np.dot(Ainv,B)
    print("Se deben transportar:")
    for i in range(len(C)):
      print(f'Cantera {i+1} = {C[i]} m^3 de material')
except np.linalg.LinAlgError:
    print("La matriz es singular y no tiene inversa")
```
