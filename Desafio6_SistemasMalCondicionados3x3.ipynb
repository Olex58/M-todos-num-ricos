```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[0.001, 1, 1],
              [1, 0.001, 1],
              [1, 1, 0.001]])
b = np.array([1, 1, 1])

det_A = np.linalg.det(A)
x = np.linalg.solve(A, b)

b_perturbado = np.array([1.001, 1, 1])
x_perturbado = np.linalg.solve(A, b_perturbado)
det_A, x, x_perturbado, x - x_perturbado

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[0], x[1], x[2], color='blue', label='Solución original', s=100)
ax.scatter(x_perturbado[0], x_perturbado[1], x_perturbado[2], color='red', label='Solución perturbada', s=100)

ax.set_title('Sistema Mal Condicionado 3x3')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()
plt.show()
```
