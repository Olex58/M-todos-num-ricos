```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n f(t) = 20 * exp(-0.1 * t)
def f(t):
    return 20 * np.exp(-0.1 * t)

# Crear un arreglo de d√≠as t (de 0 a 30)
t_values = np.linspace(0, 30, 100)

# Calcular las tasas de cambio f(t) para cada valor de t
f_values = f(t_values)

# Calcular el apoyo acumulado (integral num√©rica usando el m√©todo de integraci√≥n trapezoidal)
apoyo_acumulado = np.cumsum(f_values) * (t_values[1] - t_values[0])  # Aproximaci√≥n num√©rica

# Imprimir el apoyo acumulado en el d√≠a 30
print(f"Apoyo acumulado en 30 d√≠as: {apoyo_acumulado[-1]:.2f}")

# Graficar la tasa de cambio y el apoyo acumulado
plt.figure(figsize=(10, 6))

# Graficar la tasa de cambio f(t)
plt.subplot(2, 1, 1)
plt.plot(t_values, f_values, label="Tasa de cambio f(t)", color='blue')
plt.title("Tasa de Cambio de Apoyo en Funci√≥n del Tiempo")
plt.xlabel("D√≠a (t)")
plt.ylabel("Tasa de cambio f(t)")
plt.grid(True)
plt.legend()

# Graficar el apoyo acumulado
plt.subplot(2, 1, 2)
plt.plot(t_values, apoyo_acumulado, label="Apoyo acumulado", color='green')
plt.title("Apoyo Acumulado en Funci√≥n del Tiempo")
plt.xlabel("D√≠a (t)")
plt.ylabel("Apoyo acumulado")
plt.grid(True)
plt.legend()

# Mostrar las gr√°ficas
plt.tight_layout()
plt.show()
```
