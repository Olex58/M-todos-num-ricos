```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import hilbert, det, norm, solve

tam=range(1, 11)
deter=[]
cn=[]
for n in tam:
    H = hilbert(n)
    deter.append(det(H))
    cn.append(norm(H) * norm(np.linalg.inv(H)))
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(tam, deter, marker='o', color='b')
plt.yscale("log")
plt.title("Determinante de la matriz de Hilbert")
plt.xlabel("Tamaño de la matriz (n x n)")
plt.ylabel("Determinante")
plt.subplot(1, 2, 2)
plt.plot(tam, cn, marker='o', color='r')
plt.yscale("log")
plt.title("Número de condición de la matriz de Hilbert")
plt.xlabel("Tamaño de la matriz (n x n)")
plt.ylabel("Número de condición")
plt.tight_layout()
plt.show()
```
