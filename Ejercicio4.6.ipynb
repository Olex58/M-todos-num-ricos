```python
import math


def f(x):
    return math.log(x)

def aproximacionTaylor(x, x0):

    f_x0 = 0
    derivadaf_x0  = 1
    doblederivadaf_x0= -1
    triplederivadaf_x0 = 2
    cuadruplederivada_x0 = -6

    # Serie de Taylor hasta el cuarto orden
    taylorCuartoOrden = (
        f_x0 +
        derivadaf_x0 * (x - x0) +
        (doblederivadaf_x0 / 2) * (x - x0)**2 +
        (triplederivadaf_x0 / 6) * (x - x0)**3 +
        (cuadruplederivada_x0 / 24) * (x - x0)**4
    )

    return taylorCuartoOrden


x0 = 1
x = 2.5


taylorCuartoOrden = aproximacionTaylor(x, x0)


valoVer = f(x)

#error relativo porcentual verdadero
erp = abs((valoVer - taylorCuartoOrden) / valoVer) * 100

# Imprimir resultados
print(f"Valor aproximado usando la serie de Taylor: {taylorCuartoOrden}")
print(f"Valor exacto de ln(2.5): {valoVer}")
print(f"Error relativo porcentual verdadero: {erp:.2f}%")
```
