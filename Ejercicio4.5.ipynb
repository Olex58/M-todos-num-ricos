```python


def f(x):
    return 25 * x**3 - 6 * x**2 + 7 * x - 88


def derivadaf(x):
    return 75 * x**2 - 12 * x + 7

def doblederivadaf(x):
    return 150 * x - 12

def triplederivadaf(x):
    return 150

x0 = 1

f_x0 = f(x0)
derivadaf_x0 = derivadaf(x0)
doblederivadaf_x0 = doblederivadaf(x0)
triplederivadaf_x0 = triplederivadaf(x0)

x = 3

aproximacionTaylor = (
    f_x0 +
    derivadaf_x0* (x - x0) +(doblederivadaf_x0/ 2) * (x - x0)**2 +(triplederivadaf_x0 / 6) * (x - x0)**3
)


ValorExacto= f(x)

# Error relativo porcentual verdadero
erp = abs((ValorExacto - aproximacionTaylor) / ValorExacto) * 100


print(f"Valor aproximado usando la serie de Taylor: {aproximacionTaylor}")
print(f"Valor exacto de f(3): {ValorExacto}")
print(f"Error relativo porcentual verdadero: {erp:.4f}%")
```
