```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)


X, Y = np.meshgrid(x, y)
Z1 = 8*X + 4*Y - 11

Z2 = 4 + 2*X - 5*Y


Z3 = (7 - 2*X + Y) / 6


ax.plot_surface(X, Y, Z1, color='red', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z2, color='green', alpha=0.5, rstride=100, cstride=100)
ax.plot_surface(X, Y, Z3, color='blue', alpha=0.5, rstride=100, cstride=100)


ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

ax.set_title('Intersección de tres planos')


ax.scatter(1, 1, 1, color='black', s=100, label="Solución (1,1,1)")
ax.legend()


plt.show()
```
