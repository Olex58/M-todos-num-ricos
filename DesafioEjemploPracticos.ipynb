```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
m = 5          # Masa (kg)
g = 9.81       # Gravedad (m/s^2)
k = 0.05       # Constante (kg/m)
t0 = 0         # Tiempo inicial (s)
tf = 15        # Tiempo final (s)
v0 = 0         # Velocidad inicial (m/s)
h = 1          # Tamaño del paso (s)

# Número de pasos
N = int((tf - t0) / h) + 1

# Inicialización
t = np.linspace(t0, tf, N)  # Vector de tiempo
v = np.zeros(N)             # Vector de velocidad
v[0] = v0                   # Condición inicial

# Función de la derivada dv/dt
def f(v):
    return -g + (k/m) * v**2

# Método de Heun
for n in range(N - 1):
    v_pred = v[n] + h * f(v[n])                     # Predicción
    v[n + 1] = v[n] + (h / 2) * (f(v[n]) + f(v_pred))  # Corrección

# Graficar los resultados
plt.figure(figsize=(8, 5))
plt.plot(t, v, label="Velocidad (Heun)", color="blue", marker='o')
plt.title("Velocidad de un objeto en caída libre (Método de Heun)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.grid()
plt.legend()
plt.show()

# Imprimir las últimas velocidades calculadas
print("Últimas velocidades calculadas:")
for i in range(-5, 0):
    print(f"t = {t[i]:.1f} s, v = {v[i]:.2f} m/s")
```
